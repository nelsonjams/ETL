```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
population_file = "../Resources/Total_Population_1960_2018.csv"
```

```python
population_df = pd.read_csv(population_file)
population_df.head()
```

```python
population_df["id"] = population_df.index + 1
population_df.head()
```

```python
population_df = population_df[(population_df['Country Name'] == "Italy") or (population_df['Country Name'] == "Japan")]
print (population_df)
```

```python
gdp_file = "../Resources/World_GDP_1960_2018.csv"
gdp_df = pd.read_csv(gdp_file)
gdp_df["id"] = gdp_df.index + 1
gdp_df.head()
```

 ### Transform population DataFrame

```python
# Create a filtered dataframe from specific columns
population_cols = ["id", 
                   "Country Name", "Indicator Name",  
                   "2008","2009","2010","2011","2012","2013",
                   "2014","2015","2016","2017", "2018"]
population_transformed= population_df[population_cols].copy()

# Rename the column headers
population_transformed = population_transformed.rename(columns={"Country Name": "country_name",
                                                          "Indicator Name": "indicator_name", 
                   "2008":"YR_2008","2009":"YR_2009","2010":"YR_2010","2011":"YR_2011","2012":"YR_2012","2013":"YR_2013",
                   "2014":"YR_2014","2015":"YR_2015","2016":"YR_2016","2017":"YR_2017", "2018":"YR_2018"})

# Clean the data by dropping duplicates and setting the index
population_transformed.drop_duplicates("id", inplace=True)
population_transformed.set_index("id", inplace=True)

population_transformed.head()
```

### Transform gdp DataFrame

```python
# Create a filtered dataframe from specific columns
gdp_cols = ["id", "Country Name", "Indicator Name",  
                   "2008","2009","2010","2011","2012","2013",
                   "2014","2015","2016","2017", "2018"]

gdp_transformed= gdp_df[gdp_cols].copy()

# Rename the column headers
gdp_transformed = gdp_transformed.rename(columns={"Country Name": "country_name",
                                                          "Indicator Name": "indicator_name", 
                   "2008":"YR_2008","2009":"YR_2009","2010":"YR_2010","2011":"YR_2011","2012":"YR_2012","2013":"YR_2013",
                   "2014":"YR_2014","2015":"YR_2015","2016":"YR_2016","2017":"YR_2017","2018":"YR_2018",})

# Clean the data by dropping duplicates and setting the index
gdp_transformed.drop_duplicates("id", inplace=True)
gdp_transformed.set_index("id", inplace=True)

gdp_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:password@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
population_transformed.to_sql(name='population', con=engine, if_exists='append', index=True)
```

```python
gdp_transformed.to_sql(name='gdp', con=engine, if_exists='append', index=True)
```
